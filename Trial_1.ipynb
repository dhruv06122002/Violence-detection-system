```python
import cv2
import numpy as np
from keras.models import load_model

def process_video(video_path, model_path='model.h5', output_path='output_video.avi'):
    model = load_model(model_path)
    vs = cv2.VideoCapture(video_path)
    (W, H) = (None, None)

    frame_width = int(vs.get(3))
    frame_height = int(vs.get(4))
    fps = int(vs.get(5))
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    while True:
        grabbed, frame = vs.read()

        if not grabbed:
            break

        if W is None or H is None:
            (H, W) = frame.shape[:2]

        output = frame.copy()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (128, 128)).astype("float32")
        frame = frame.reshape(128, 128, 3) / 255

        preds = model.predict(np.expand_dims(frame, axis=0))[0]
        violence_prob = preds[0]  # Access the first index

        if violence_prob > 0.5:
            label = "True"
            color = (0, 0, 255)  # Red
        else:
            label = "False"
            color = (0, 255, 0)  # Green

        cv2.putText(output, f"Violence: {label}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

        out.write(output)

        key = cv2.waitKey(1) & 0xFF
        if key == ord("q"):
            break

    vs.release()
    out.release()
    cv2.destroyAllWindows()

video_path = "t1.mp4"
output_video_path = "output_video.avi"
process_video(video_path, output_path=output_video_path)
```
